# 11. Reshape dataframe using melt

> Pandas melt function provides a way to transform and reshape dataframe

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/pandas/11_melt/weather.csv')
df
```

```python
df_melted = pd.melt(df,id_vars=['day'],var_name='city',value_name='temp')
df_melted
```
