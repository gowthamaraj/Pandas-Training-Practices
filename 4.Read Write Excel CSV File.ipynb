## Skip rows in dataframe using "skiprows" 

```python
import pandas as pd
df = pd.read_csv("weatherdata.csv",skiprows=1,header=None)
df
```

```python
df = pd.read_csv("weatherdata.csv",skiprows=1,header=None,names=['col1','col2','col3','col4'])
df
```

## Read limited data from CSV file

```python
df = pd.read_csv("weatherdata.csv",nrows=3)
df
```

## Clean up messy data from file "not available" and "n.a." replace with "na_values"

```python
df = pd.read_csv("weatherdata.csv",na_values=['Rain'])
df
```

```python
df = pd.read_csv("weatherdata.csv",na_values={'event':['Rain']})
df
```
