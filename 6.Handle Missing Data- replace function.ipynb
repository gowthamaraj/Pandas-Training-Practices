# 6. Handle Missing Data: replace function

```python
import pandas as pd
import numpy as np
url = "https://raw.githubusercontent.com/codebasics/py/master/pandas/6_handling_missing_data_replace/weather_data.csv"
df = pd.read_csv(url)
df
```

##  Use replace() method to replace values in dataframe

```python
new_df = df.replace(-99999,np.NaN)
new_df
```

##  Replace values using a dictionary

```python
new_df = df.replace({'temperature':-99999,'event':'0'},np.NaN)
new_df
```

```python
new_df = df.replace({-99999:np.NaN,'Rain':'Rainy'})
new_df
```

## How "regex" (regular expression) works 

```python
df_regex = df.replace('[A-Za-z]','',regex=True)
df_regex
```

```python
df_regex = df.replace({'temperature':'[A-Za-z]','windspeed':'[A-Za-z]'},'',regex=True)
df_regex
```
