# 7. Group By (Split Apply Combine)

> In this python pandas tutorial you will learn how groupby method can be used to group your dataset based on some criteria and then apply analytics on each of the groups. This is similar to SQL group by. It is also called split apply combine strategy in data science.

```python
import pandas as pd
url = "https://raw.githubusercontent.com/codebasics/py/master/pandas/7_group_by/weather_by_cities.csv"
df = pd.read_csv(url)
df
```

***

## Use groupby() method 

```python
df_groupby = df.groupby('city')
df_groupby
```

Print out the pandas.core.groupby.generic.DataFrameGroupBy

```python
for city,city_df in df_groupby:
    print(city)
    print(city_df)
```

[or]

```python
df_groupby.get_group('paris')
```

```python
df_groupby.max()
```

```python
df_groupby.mean()
```

## Use describe() function in groupby

```python
df_groupby.describe()
```
