# 8. Concat Dataframes

```python
import pandas as pd

india_weather = pd.DataFrame({
    'city':['chennai','bangalore','delhi'],
    'weather':[15,20,14],
    'humidity':[60,40,39]
})

india_weather
```

```python
usa_weather = pd.DataFrame({
    'city':['nyu','california','dc'],
    'weather':[25,40,30],
    'humidity':[70,50,42]
})

usa_weather
```

```python
new_df = pd.concat([usa_weather,india_weather])
new_df
```

```python
new_df = pd.concat([usa_weather,india_weather],ignore_index=True)
new_df
```

##  What is "keys"? pass "keys" to concat() function

```python
new_df = pd.concat([usa_weather,india_weather],keys=['usa','india'])
new_df
```

```python
new_df.loc['india']
```

## "axis" argument in concat() function

```python
temp_df = pd.DataFrame({
    'city':['nyu','california','dc'],
    'weather':[25,40,30]
})
temp_df
```

```python
hum_df = pd.DataFrame({
    'city':['nyu','california','dc'],
    'humidity':[70,50,42]
})
hum_df
```

```python
new_df = pd.concat([hum_df,temp_df],axis=1)
new_df
```

## Join dataframe with series() function 

```python
ser = pd.Series(['a','b','c'],name='rank')
ser
```

```python
new_df = pd.concat([hum_df,ser],axis=1)
new_df
```
