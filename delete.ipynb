```python
import pandas as pd
import numpy as np
import locale
import warnings
from locale import atof
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel("Master Finance Data.xlsx", thousands=',')
df1 = df.drop(['Student Name'], axis=1)
```

```python
missing_cols, missing_rows = (
    (df1.isnull().sum(x) | df1.eq('').sum(x))
    .loc[lambda x: x.gt(0)].index
    for x in (0, 1)
)
null_df = df1.loc[missing_rows, missing_cols]
```

```python
df2 = df1.dropna(how='any')
df2['Registration Date'] = pd.to_datetime(df2['Registration Date'], errors='coerce')
df2.replace(',','', regex=True, inplace=True)
cols = ['Total Fee','Paid', 'Due']
df2[cols] = df2[cols].applymap(lambda x: x if not isinstance(x, str) else np.nan)
df2.fillna(0, inplace=True)
df2['Total Fee'].replace('-','', regex=True, inplace=True)
final_df = df2[(df2['Total Fee']>=0)|(df2['Paid']>=0)]
final_df
```

```python
cols = ['Total Fee', 'Paid','Due']
final_df[cols] = final_df[cols].applymap(np.int64)
final_df['Month/Year'] = df['Month/Year'].dt.strftime('%m-%y')
final_df
```
